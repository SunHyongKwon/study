### 예제1 : 괄호맞추기  
- (()()) 이처럼 쌍이 맞으면 True , (()))( 이처럼 안맞으면 False를 반환하는 함수 작성하기
 

```python
from data_structure.stack import Stack

def is_valid_parentheses(s : str) -> bool:
    stack = Stack()
    for char in s:
        if char == '(':
            stack.push(char)
        elif char == ')':
            if stack.pop() != '(':
                return False
    
    return stack.size() == 0

print(is_valid_parentheses("(()())"))
print(is_valid_parentheses("(()))("))
```

#### 예제2 : 계산기 코드 작성
- 2 + 3 * 5 와 같은 수식을 받으면 이를 계산하는 예제
- 연산자 별 우선순위가 다르고 지금 infix로 되어 있는 것을 postfix로 변환한 후 이를 연산하는 과정을 실습하는 것임

```python
from data_structure.stack import Stack
import re

def infix_to_postfix(exp : str) -> Stack:
    # 수식 문자열을 리스트로 변경 하여 2자리 수 이상의 숫자도 분리되도록 함
    exp = re.split(r'(\+|-|\*|/|\(|\))', exp)
    exp = [op.strip() for op in exp if op.strip() != ""]

    postfix_stack = Stack()
    operator_stack = Stack()

    operator_priority = {'+': 1,'-': 1,'*': 2,'/': 2}

    for op in exp:
        # 1. 숫자인지 체크
        if op.isdigit():
            postfix_stack.push(op)
        # 2. 연산자인지 체크
        elif op in ['+', '-', '*', '/']:
            while operator_stack.size() > 0 \
                and operator_stack.top() != '(' \
                and operator_priority[op] <= operator_priority[operator_stack.top()]:
                postfix_stack.push(operator_stack.pop())
            operator_stack.push(op)
        # 3. 여는 괄호인지 체크
        elif op == '(':
            operator_stack.push(op)
        # 4. 닫는 괄호인지 체크
        elif op == ')':
            while operator_stack.top() != '(':
                postfix_stack.push(operator_stack.pop())
            operator_stack.pop()

    # 남은 연산자들 postfix_stack에 추가
    while operator_stack.size() > 0:
        postfix_stack.push(operator_stack.pop())

    return postfix_stack


def calculate_postfix(exp : str) -> int:
    postfix_stack = infix_to_postfix(exp)
    number_stack = Stack()
    result = 0
    for op in postfix_stack.items:
        if op.isdigit():
            number_stack.push(int(op))
        else:
            a = number_stack.pop()
            b = number_stack.pop()
            if op == '+':
                result = b + a
            elif op == '-':
                result = b - a
            elif op == '*':
                result = b * a
            elif op == '/':
                result = b / a
            number_stack.push(result)
    return result
```

```python
calculate_postfix("2 + 3 * 5")
```
