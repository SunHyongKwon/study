#### 예제 1: Josephus Problem
- n과 k가 주어지고 n의 인원이 동그랗게 앉아서 k번째 사람마다 사라지는 것을 가정했을 때 마지막에 남는 숫자는 무엇일까

```python
from data_structure.queue import Queue

def josephus(n: int, k: int) -> int:
    queue = Queue()

    for i in range(1, n+1):
        queue.enqueue(i)

    while queue.size() > 1:
        for _ in range(k-1):
            queue.enqueue(queue.dequeue())
        queue.dequeue()

    return queue.dequeue()

josephus(9, 3)
```
