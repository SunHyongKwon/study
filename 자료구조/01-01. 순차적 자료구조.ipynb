### 1. 순차적 자료구조 소개

1. 배열, 리스트  
&emsp; - index로 임의의 원소를 접근  
<br>
2. Stack, Queue, Dequeue  
&emsp; - 제한된 접근(삽입,삭제)만 허용  
&emsp; - Stack : LIFO -> 맨 뒤로 들어오고 맨 뒤에 들어온 값이 나가는 것  
&emsp; - Queue : FIFO -> 맨 뒤로 들어오고 맨 처음 들어온 값이 나가는 것  
&emsp; - Dequeue : Stack 과 Queue를 합쳐놓은 거고 맨 뒤나 맨 처음으로 들어올 수 있고 맨 뒤나 맨처음 값이 나가는 것  
<br>
3. Linked List(연결 리스트)  
&emsp; - 각각의 값들과 다음 값의 주소를 가지는 구조  
&emsp; - 인덱스로 접근 할 수 없지만, 삽입 삭제의 시간 복잡도가 좋음

### 2. C의 배열과 Python의 리스트의 차이점

1. C의 배열  
&emsp; - C의 경우에는 길이가 정해져 있고, 메모리의 순차적으로 들어가는 거임  
&emsp; - 예를 들어서 int[4] 의 배열이 있다고 하면 100번 메모리에 0번 인덱스가 들어가면 그 byte만큼 뒤에 104번 메모리에 다음 숫자가 들어오는 구조  
<br>

2. Python의 리스트  
&emsp; - python의 경우에는 숫자들은 메모리 상에 흩어져 있고 리스트는 이것의 포인터를 보는 구조  
&emsp; -  그리고 리스트의 크기를 지정하지 않아도 자동으로 증가하게 되는 구조를 가지고 있음
