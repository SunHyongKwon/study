### 예제1 : 괄호맞추기  
- (()()) 이처럼 쌍이 맞으면 True , (()))( 이처럼 안맞으면 False를 반환하는 함수 작성하기
 

```python
from data_structure.stack import Stack

def is_valid_parentheses(s):
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if stack.pop() != '(':
                return False
            else:
                return True
    return len(stack) == 0

print(is_valid_parentheses("(()())"))
print(is_valid_parentheses("(()))("))
```
